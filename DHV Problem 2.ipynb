```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Replace the URL or path with the actual location of the Parquet file
file_url = 'exercises_problem2.parquet'
data = pd.read_parquet(file_url)
```

```python
# Display the first few rows of the dataset
print(data.head())

# Summary statistics for numerical columns
print(data.describe())

# Data types of each column
print(data.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['var1'], kde=True, color='gray')
plt.title('Distribution of Var1')
plt.xlabel('Var1')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='var1', y='var2', hue='class', style='class', palette='coolwarm')
plt.title('Relationship between Var1 and Var2 with Class Differentiation')
plt.xlabel('Var1')
plt.ylabel('Var2')
plt.legend(title='Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='var2', y='var1', hue='class', data=data, palette='coolwarm')
plt.title('Var1 Over Var2 by Class')
plt.xlabel('Var2')
plt.ylabel('Var1')
plt.legend(title='Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='class', y='var1', data=data, palette='coolwarm', ci=None)
plt.title('Average Var1 by Class')
plt.xlabel('Class')
plt.ylabel('Average Var1')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='var1', data=data, palette='coolwarm')
plt.title('Var1 Distribution across Classes')
plt.xlabel('Class')
plt.ylabel('Var1')
plt.show()
```
