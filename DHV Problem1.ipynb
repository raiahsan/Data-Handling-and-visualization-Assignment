```python
import matplotlib.pyplot as plt
import numpy as np

# Points
x1, y1 = 3, 5
x2, y2 = 7, 2

# Now Calculate the slope (s)m and y-intercept (a)b of the line
s = (y2 - y1) / (x2 - x1)
a = y1 - s * x1

# Generate x values for the line plot extending to the edges
x_values = np.linspace(1, 9, 100)

# Calculate corresponding y values based on the line equation
y_values = s * x_values + a

# Plotting
plt.figure(figsize=(8, 6))

# Plot the line
plt.plot(x_values, y_values, label='Line', color='black', linewidth=2)

# Plot the points
plt.scatter([x1, x2], [y1, y2], color='black', s=200, zorder=20)

# Styling
plt.xlim(1, 9)
plt.ylim(0, 6)
plt.xlabel('X axis', fontsize=10)
plt.ylabel('Y axis', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='black', linewidth=2)
plt.axvline(x=0, color='black', linewidth=2)

# Display the plot
plt.show()
```
